# Hands-on Practice — DataFrame Operations

## Objective

Create a dataset with the following columns:

- name  
- marks  
- city  

Then perform:

- Compute average marks  
- Add grade column  
- Filter students above 80  
- Print summary statistics  
- Add bonus marks column  


```python
import pandas as pd

data={"name":["Ayush","Rahul","Raj"],
      "age":[23,25,27],
      "marks":[90,75,60]}

df=pd.DataFrame(data)
print(df)
```

```python
# compute average marks
avg_marks=df["marks"].mean()
print("Average marks: ",avg_marks)
```

```python
# add grade column: 90+ → A, 80–89 → B, 70–79 → C, below 70 → D
def assign_grade(mark):
  if mark>=90:
    return "A"
  elif mark>=80:
    return "B"
  elif mark>=70:
    return "C"
  else:
    return "D"

df["grade"]=df["marks"].apply(assign_grade)   # .apply() applies a function to every row in a column.
print(df)
```

```python
# filter students above 80 marks
print(df[df["marks"]>80])

# df[ something ]
# something must be one of these:
# A column name → df["marks"]
# A list of column names → df[["marks", "name"]]
# A Boolean condition (True/False values) → df[df["marks"] > 80]
```

```python
# Print summary statistics
print(df.describe())
```

```python
# Add Bonus Marks Column
df["bonus_marks"] = df["marks"] + 5
print(df)
```

# HANDS-ON — Mini Data Project

## Scenario

You work as a Data Analyst for a coaching institute.

You are given:

- Student performance data  
- City-level bonus marks data  

You must:

- Handle missing values  
- Sort data  
- Perform group analysis  
- Merge datasets  
- Generate insights  

```python
# STEP 1 — Create Student Dataset (with missing values)
import pandas as pd
import numpy as np

students_data = {
    "student_id": [1, 2, 3, 4, 5, 6],
    "name": ["Ayush", "Riya", "Arjun", "Neha", "Kabir", "Meera"],
    "marks": [85, 92, np.nan, 88, 67, 76],
    "city": ["Delhi", "Mumbai", "Delhi", "Bangalore", "Mumbai", None]
}

df_students = pd.DataFrame(students_data)

print(df_students)
```

```python
# STEP 2 — Handle Missing Values

# Check missing values
print(df_students.isnull().sum())    # df_students.isnull() Returns a DataFrame of True / False. True → value is missing (NaN). False → value is present. Since in Python: True = 1, False = 0 .sum() counts how many True values are in each column.

# Fill missing marks with average
avg_mark=df_students["marks"].mean()
df_students["marks"]=df_students["marks"].fillna(avg_mark)   # df_students["marks"].fillna(avg_marks, inplace=True) can also be written like this but this is older way. df["marks"]=df["marks"].fillna(avg_mark) is ,odern way

# Fill missing city with "Unknown"
df_students["city"]=df_students["city"].fillna("Unknown")   # can also do df_students=df_students.fillna({"city":"Unknown"}) but df_students["city"]=df_students["city"].fillna("Unknown") is better.

# Verify
print(df_students.isnull().sum())
print(df_students)
```

```python
# STEP 3 — Add Status Columns
df_students["status"]=df_students["marks"].apply(
    lambda x: "PASS" if x>=75 else "FAIL"
)
print(df_students)
```

```python
# STEP 4 — Sorting

# Sort by marks (highest first)
sorted_df=df_students.sort_values(by="marks",ascending=False)  # by default is asc=true. NOTE: sort_values() does not change the original DataFrame unless Option 1 (recommended): df_students = df_students.sort_values(by="marks", ascending=False) or Option 2: df_students.sort_values(by="marks", ascending=False, inplace=True)
print(sorted_df)

# Sort by city then marks
sorted_city=df_students.sort_values(by=["city","marks"])  # sort by city and if city is same sort by marks
print(sorted_city)
```

```python
# STEP 5 — GroupBy Deep Dive

# Average marks per city
city_avg = df_students.groupby("city")["marks"].mean()
print(city_avg)

# Count students per city
city_count = df_students.groupby("city").size()
print(city_count)

# Multiple aggregations
city_stats = df_students.groupby("city")["marks"].agg(["mean", "max", "min", "count"])
print(city_stats)

# Multiple columns
multiCol= df_students.groupby("city")[["marks", "student_id"]].agg(["mean", "max"])
print(multiCol)

# Custom function
customFunc=df_students.groupby("city")["marks"].agg(["mean", lambda x: x.std()])
print(customFunc)

```

```python
# STEP 6 — Create Bonus Dataset (Second Table)
bonus_data = {
    "city": ["Delhi", "Mumbai", "Bangalore"],
    "bonus_marks": [5, 3, 4]
}

df_bonus = pd.DataFrame(bonus_data)

print(df_bonus)
```

```python
# STEP 7 — Merge Datasets

# pd.merge() This function combines two DataFrames based on a common key (column). Think of it like SQL JOIN operations.
# df_students → the “main” table. df_bonus → the table containing extra info per city
# on="city" → join key
# how="left" Specifies the type of join (like SQL). left → keep all rows from the left DataFrame (df_students)

merged_df=pd.merge(df_students,df_bonus,on="city",how="left")
print(merged_df)
```

```python
# STEP 8 — Final Feature Engineering
merged_df["final_marks"] = merged_df["marks"] + merged_df["bonus_marks"]
print(merged_df)
```

```python
# STEP 9 — Real Analysis Questions

# Top performer after bonus
top=merged_df.sort_values(by="final_marks",ascending=False).head(1)
print(top)

# Which city performs best on average (after bonus)
topCity=merged_df.groupby("city")["final_marks"].mean()
print(topCity)

# How many students passed
pass_count = merged_df["status"].value_counts()
print(pass_count)
# if want only pass students count
pass_count = (merged_df["status"] == "PASS").sum()
print(pass_count)
```

# MINI PROJECT

## Scenario

You are analyzing a fictional Pokémon-style dataset for a gaming analytics company.

You must:

- Clean the dataset  
- Standardize values  
- Merge bonus data  
- Perform city-level and type-level analysis  
- Generate insights  

```python
# STEP 1 — Create Raw Dataset (Messy On Purpose)

import pandas as pd
import numpy as np

data = {
    "No": [1, 2, 3, 4, 5, 5],
    "Name": ["Bulbasaur", "Charmander", "Squirtle", "Pikachu", "Eevee", "Eevee"],
    "Type1": ["Grass", "Fire", "Water", "Electric", "Normal", "Normal"],
    "Type2": [np.nan, np.nan, np.nan, np.nan, "Fairy", "Fairy"],
    "Total": [318, 309, 314, 320, 325, 325],
    "City": ["Delhi", "Mumbai", "Delhi", None, "Mumbai", "Mumbai"],
    "Legendary": [False, False, False, False, False, False]
}

df = pd.DataFrame(data)

print(df)
```

```python
# STEP 2 — Drop Unnecessary Columns

df = df.drop(columns=["No"])  # No i.e., Number col will be removed
print(df)
```

```python
# STEP 3 — Handle Missing Data

# Check missing
print(df.isnull().sum())

# Option A — Drop rows where Type2 is missing
df_drop = df.dropna(subset=["Type2"])
print(df_drop)

# Option B — Fill missing Type2 with "None"
df = df.fillna({"Type2": "None"})
print(df)

# Fill missing City with "Unknown"
df["City"] = df["City"].fillna("Unknown")
# df = df.fillna({"City":"Unknown"})  # can also do multiple cols: df = df.fillna({"City": "Unknown","marks": df["marks"].mean()})

print(df)
```

```python
# STEP 4 — Fix Inconsistent Values
df["Type1"]=df["Type1"].replace({"Grass":"GRASS","Fire":"FIRE"})
print(df)
```

```python
# STEP 5 — Standardize Text
df["Name"] = df["Name"].str.lower()
print(df)
```

```python
# STEP 6 — Fix Data Types
df["Legendary"] = df["Legendary"].astype(bool)
print(df)
```

```python
# STEP 7 — Remove Duplicate Rows
df = df.drop_duplicates()
print(df)
```
