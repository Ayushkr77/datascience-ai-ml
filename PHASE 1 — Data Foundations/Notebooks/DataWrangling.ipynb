# Hands-On Practice (Critical) — Data Wrangling

## Objective

Create a dataset with:

- employee_id  
- department  
- city  
- salary  

Then perform:

- Average salary per department  
- Total salary per city  
- Highest salary per department  
- Create tax column  
- Filter employees earning > 60k  
- Create another table (bonus info)  
- Merge both tables  
- Create pivot table  

```python
# STEP 1 — Create Dataset

import pandas as pd

data = {
    "employee_id": [101, 102, 103, 104, 105, 106],
    "department": ["IT", "HR", "IT", "Finance", "HR", "Finance"],
    "city": ["Delhi", "Mumbai", "Delhi", "Bangalore", "Mumbai", "Delhi"],
    "salary": [65000, 48000, 72000, 90000, 52000, 85000]
}

df = pd.DataFrame(data)

print("Original Dataset:")
print(df)
```

```python
# Average Salary per Department
avgSal=df.groupby("department")["salary"].mean()
print(avgSal)
```

```python
# Total Salary per City
totSal=df.groupby("city")["salary"].sum()
print(totSal)
```

```python
# Highest Salary per Department
highestSal=df.groupby("department")["salary"].max()
print(highestSal)
```

```python
# Create Tax Column (15%)
df["tax"]=df["salary"]*0.15
print(df)
```

```python
# Filter Employees Earning > 60k
highEarners=df[df["salary"]>60000]
print(highEarners)
```

```python
# Create Bonus Table (Second Table)

bonus_data = {
    "employee_id": [101, 102, 103, 104, 105, 106],
    "bonus": [5000, 3000, 7000, 10000, 4000, 8000]
}

df_bonus = pd.DataFrame(bonus_data)

print("\nBonus Table:")
print(df_bonus)
```

```python
# Merge Both Tables

merged_df=pd.merge(df,df_bonus,on="employee_id",how="left")
print(merged_df)
```

```python
# Create Pivot Table
# Average salary per department per city:

pivot_table = merged_df.pivot_table(
    values="salary",
    index="department",
    columns="city",
    aggfunc="mean"
)

print("\nPivot Table (Average Salary per Department per City):")
print(pivot_table)



# pivot table = groupby + reshape + aggregate. also see diff bet groupby and pivot

# pivot_table Syntax
# df.pivot_table(
#     index="row_column",       # rows of the new table
#     columns="column_column",  # columns of the new table
#     values="value_column",    # data to aggregate
#     aggfunc="sum"             # aggregation function, default is 'mean'
# )

# Total Sales per Region per Product
# pivot = df.pivot_table(
#     index="Region",
#     columns="Product",
#     values="Sales",
#     aggfunc="sum"
# )
# print(pivot)
# | Product | A   | B   |
# | ------- | --- | --- |
# | East    | 150 | 150 |
# | West    | 200 | 250 |

# suppose original table is this, now see above output
# | Index | Region | Product | Sales |
# | ----- | ------ | ------- | ----- |
# | 0     | East   | A       | 100   |
# | 1     | East   | B       | 150   |
# | 2     | West   | A       | 200   |
# | 3     | West   | B       | 250   |
# | 4     | East   | A       | 50    |

```
