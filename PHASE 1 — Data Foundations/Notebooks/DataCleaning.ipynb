# Hands-on Practice — Data Cleaning

## Objective

Create a fake messy dataset containing:

- Duplicate rows  
- Missing ages  
- City with inconsistent text  
- Salary stored as string  

Then:

- Clean missing values  
- Remove duplicates  
- Fix text  
- Convert salary to float  
- Confirm dataset is clean  
- Print before and after  


```python
# STEP 1 — Create a Messy Dataset

import pandas as pd
import numpy as np

data = {
    "name": ["Ayush", "Riya", "Arjun", "Neha", "Kabir", "Kabir"],
    "age": [22, np.nan, 25, 23, 24, 24],   # missing age + duplicate row
    "salary": ["50000", "60000", "55000", "52000", "9999999", "9999999"],  # salary as string + extreme value
    "city": [" Delhi", "mumbai ", "DELHI", " Bangalore ", "Mumbai", "Mumbai"],
    "joining_date": ["2022-01-15", "2021-07-10", "2023-03-12",
                     "2022-11-01", "2020-05-20", "2020-05-20"]
}

df = pd.DataFrame(data)

print("Raw Data:")
print(df)

# Problems intentionally added:
# Duplicate row (Kabir)
# Missing age
# Salary stored as string
# One extreme salary value (9999999)
# Inconsistent city formatting
# Date stored as string
```

```python
# STEP 2 — Remove Duplicates

print("Number of duplicate rows before removing:", df.duplicated().sum())

df = df.drop_duplicates()

print("Number of duplicate rows after removing:", df.duplicated().sum())

print("\nAfter Removing Duplicates:")
print(df)
```

```python
# STEP 3 — Fix Data Types

# Convert salary to numeric
df["salary"] = df["salary"].astype(float)

# Convert joining_date to datetime
df["joining_date"] = pd.to_datetime(df["joining_date"])

# Check:
print(df.dtypes)
```

```python
# STEP 4 — Handle Missing Age

df["age"] = df["age"].fillna(df["age"].mean())

print("\nAfter Filling Missing Age:")
print(df)
```

```python
# STEP 5 — Clean City Column

df["city"] = df["city"].str.strip().str.lower()  # str.strip() → removes leading and trailing spaces.

print("\nAfter Cleaning City:")
print(df)
```

```python
# STEP 6 — Remove Salary Outlier (IQR Method) Interquartile Range

# Calculate IQR
Q1 = df["salary"].quantile(0.25)
Q3 = df["salary"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter Outliers
df = df[(df["salary"] >= lower_bound) & (df["salary"] <= upper_bound)]

print("\nAfter Removing Outliers:")
print(df)

# The extreme 9,999,999 salary will be removed.



# IQR Explanation

# Outlier      Whisker     Box (IQR)     Whisker     Outlier
#    o          |--------- [=====|=====] ---------|         o
#              Min        Q1  Median  Q3        Max

# 1. Calculate IQR:
#    IQR = Q3 − Q1

# 2. Lower Whisker:
#    Lower whisker = Q1 − 1.5 × IQR
#    → But it cannot go below the minimum data point.

# 3. Upper Whisker:
#    Upper whisker = Q3 + 1.5 × IQR
#    → But it cannot go above the maximum data point.

# 4. Outliers:
#    Any data points beyond the whiskers are considered outliers and often plotted as dots.


# Important Analytical Thinking
# Before removing outlier, always ask:
# Is it data entry error?
# Is it rare but real (e.g., CEO salary)?
# Will removing it bias results?
# Never remove blindly.
```

```python
# FINAL CLEANED DATASET

print("\nFinal Cleaned Dataset:")
print(df)

print("\nDataset Info:")
print(df.info())
```
