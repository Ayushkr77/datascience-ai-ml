# Matplotlib — Basic Plotting Functions Reference

```python
import matplotlib.pyplot as plt  # Imports pyplot module for plotting graphs
```

---

# 📊 Basic Plotting Functions

## 1️⃣ Line Plot

```python
plt.plot(x, y)  # Creates a line graph connecting x and y values
```

---

## 2️⃣ Bar Chart

```python
plt.bar(x, height)   # Creates a vertical bar chart
plt.barh(y, width)   # Creates a horizontal bar chart
```

---

## 3️⃣ Scatter Plot

```python
plt.scatter(x, y)  # Creates a scatter plot to show relationship between variables
```

---

## 4️⃣ Histogram

```python
plt.hist(data, bins=10)  # Creates a histogram to show data distribution
```

---

## 5️⃣ Pie Chart

```python
plt.pie(values, labels=labels)  # Creates a pie chart to show proportions
```

---

## 6️⃣ Box Plot

```python
plt.boxplot(data)  # Creates a box-and-whisker plot to show data distribution
```

---


# 📝 Titles and Labels

```python
plt.title("Title")      # Adds a title to the plot
plt.xlabel("X-axis")    # Labels the x-axis
plt.ylabel("Y-axis")    # Labels the y-axis
```

---

# 🎨 Styling Functions

```python
plt.colorbar()              # Adds a color scale bar to the plot
plt.legend()                # Displays legend for labeled data
plt.grid(True)              # Adds grid lines to the plot
plt.xlim(min, max)          # Sets limits for x-axis
plt.ylim(min, max)          # Sets limits for y-axis
plt.xticks()                # Sets or gets x-axis tick marks
plt.yticks()                # Sets or gets y-axis tick marks
plt.style.use("ggplot")     # Applies a predefined style to the plot
```

---

# 📐 Figure Control

```python
plt.figure(figsize=(8,5))   # Creates a new figure with specified size
plt.subplot(1,2,1)          # Creates a subplot in a grid layout
plt.tight_layout()          # Adjusts spacing between subplots automatically
```

---

# 💾 Output Functions

```python
plt.show()                  # Displays the plot window
plt.savefig("plot.png")     # Saves the plot to a file
plt.close()                 # Closes the current figure
```

## Line Plot — Trends Over Time

````python
import matplotlib.pyplot as plt

months = [1, 2, 3, 4, 5]
sales = [100, 150, 200, 180, 220]

plt.plot(months, sales)
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
````

## Bar Chart — Compare Categories

````python
categories = ["IT","HR","Finance"]
avg_salary = [70000,50000,65000]

plt.bar(categories, avg_salary)
plt.title("Average Salary by Department")
plt.show()
````

## Histogram — Distribution

````python
import numpy as np

data = np.random.normal(50000, 8000, 1000)   # 50000 → Mean (average salary = $50,000), 8000 → Standard deviation (how spread out salaries are), 1000 → Number of samples (1,000 employees)
# So you're simulating salaries centered around $50,000, with most values falling roughly between: $42,000 and $58,000 (within 1 standard deviation)

plt.hist(data, bins=20)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Number of persons")
plt.show()
````

## Scatter Plot — Relationships

````python
# matplotlib version

import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    "experience":[1,2,3,4,5],
    "salary":[30000,40000,50000,60000,75000]
})

plt.scatter(df["experience"], df["salary"], color="b")  # color is optional
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Matplotlib - plt.scatter()")
plt.show()
````

````python
# Matplotlib → uses arrays (df["experience"])
# Seaborn → uses column names ("experience") + data=df

# Seaborn Version
import seaborn as sns   # for statistical graphs (built on matplotlib)
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    "experience":[1,2,3,4,5],
    "salary":[30000,40000,50000,60000,75000]
})

sns.scatterplot(x="experience", y="salary", data=df, color="red")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Seaborn - sns.scatterplot()")
plt.show()
````

## Boxplot — Outlier Detection

````python
sns.boxplot(x=df["salary"])
plt.show()

# A boxplot (box-and-whisker plot) shows:
# Median
# Spread of data
# Outliers
# Minimum & Maximum
# It summarizes distribution in a compact way.

# Whiskers → Min & Max (excluding outliers)
````

## Pie Chart

````python
import matplotlib.pyplot as plt

# Sample data
departments = ["IT", "HR", "Sales", "Marketing"]
num_employees = [10, 5, 8, 7]

# Create pie chart
plt.pie(num_employees, labels=departments, autopct="%1.1f%%", startangle=90)
plt.title("Employees by Department")
plt.show()
````
